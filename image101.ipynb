# learn how to maniplate the image

```python
from keras.preprocessing import image
import numpy as np
import os
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from scipy.misc import imread, imresize
%matplotlib inline 
```

```python
data_dir = '/Users/xuan/work/dataset/cat-no-cat/img'
```

```python
def print_typeshape(img):
    print("Type is %s" % (type(img)))
    print("Shape is %s" % (img.shape,))
```

```python
cat1 = imread(data_dir + "/cat/test.jpg")
#.astype(np.float)
print_typeshape (cat1)
# plot
plt.figure(0)
plt.imshow(cat1)
plt.title("Original Image with imread")
plt.draw()
```

```python
datagen = ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization=True,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)
data_generator = datagen.flow_from_directory( data_dir, target_size=(224, 224), batch_size=1, class_mode='binary')

def next_one():
    imgs = data_generator.next()[0]
    return imgs[0].astype(np.uint8)

img = next_one()
print_typeshape (img)

plt.figure(figsize=(16,16))
SZ = 8
for i in range(SZ):
    plt.subplot(1,SZ,i+1)
    plt.imshow(next_one())
plt.draw()
```
