# use keras to fit the simple data

```python
import data
import model
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import keras
from keras.models import Model, Sequential, Input
from keras.layers import Dense
%matplotlib inline
reload(data)
reload(model)
```

```python
# generate the train and val data
m=64
reload(data)
def sample(size):
    X = np.random.random(size) * 2.0 - 1.0
    noise = np.random.random(size) / 10.0
    Y = X * X + 1.0 + noise
    return X.reshape(size, 1),Y.reshape(size, 1)

train_X, train_y = sample(m)
val_X, val_y = sample(m/3)

#plot the sample and val data
plt.plot(train_X, train_y, "bo")
plt.plot(val_X, val_y, "ro")
print "train X shape", train_X.shape
print "train y shape", train_y.shape
```

```python
# build the keras model
x = Input(shape=(1,))
y = Dense(2, activation='relu')(x)
y = Dense(1, activation='relu')(y)
model = Model(x, y)
model.summary()
```

```python
model.compile(optimizer='sgd',
              loss='mse')
model.fit(train_X, train_y, epochs=1000, batch_size=32, shuffle=True)
```

```python
# cost = model.evaluate(train_X, train_y, batch_size=32)
# print cost
# plt.figure()
y_ = model.predict(train_X)

plt.plot(train_X, train_y, "go")
plt.plot(train_X, y_, "ro")
plt.show()
```
